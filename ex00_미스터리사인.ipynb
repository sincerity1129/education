# Question 1
### 큰 수 + 두수의 차
### 22 ? 7 = 37

```python
list1 = []
for i in range(1,200):
    for j in range(1,200):
        result = (i+(i-j)) if i > j else (j + j-i)
        list1.append([i,j,result])
```

```python
list1
```

```python
data1 = pd.DataFrame(list1, columns = ['N1', 'N2', 'Result'])
data1.head()
```

```python
#문제와 정답을 나눈 코드
X = data1. loc[:, ['N1', 'N2']]
y = data1.loc[:, 'Result']
```

```python
# 머신러닝 모델 다운로드
from sklearn.ensemble import RandomForestRegressor

# 변수에 모델 담기
random_forest = RandomForestRegressor()

# 모델에 문제와 정답 학습시키기
random_forest.fit(X,y)

# 정해진 규칙을 통해 새로운 데이터에 대해 예측값 출력
random_forest.predict([[22,7]])
```

# Question2
- ( 큰 수/작은 수)의 나머지
- 4 ? 19 = 3

```python
list2 = []
for i in range(1, 200):
    for j in range(1, 200):
        result2 = (i%j) if i > j else (j%i)
        list2.append([i,j,result2])
```

```python
list2
```

```python
data2 = pd.DataFrame(list2, columns = ['N1', 'N2', "result"])
data2.head()
```

```python
X = data2.loc[:, ['N1','N2']]
y = data2.loc[:,'result']
```

```python
from sklearn.ensemble import RandomForestRegressor

random_forest = RandomForestRegressor()

random_forest.fit(X,y)

random_forest.predict([[4,19]])
```
