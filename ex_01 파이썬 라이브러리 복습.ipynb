### 조건별로 출력하기
- 면적이 130 이상인 아파트만 출력하기
- 불리언 인덱싱 사용

```python
bol = df['면적'] >= 130
price = df['가격'] <= 15000
```

```python
# 특정 데이터 선택 후 값을 뽑아 낼 때
df.loc[ bol & price ,'가격']
```

- 면적이 130이상인 아파트 가격만 출력

```python
df.loc[bol,'가격']
```

- 면적이 130이상이고 가격이 1.5억 미만인 아파트 가격만 출력

```python
df.loc[bol&price , ['가격', '면적']]
```

- 위의 3가지 조건을 모두 만족하는 아파트들의 가격의 합계를 출력

```python
sum(df.loc[(df['면적']>=130) & (df['가격']<15000) , '가격'])
```

###  원하는 값 바꾸기
- 가격이 800 미만인 값들을 0으로 바꾸기

```python
# 이렇게 담을 경우 값을 변화시키더라도 원본에는 변화가 되지 않음
# 따라서 loc를 활용하여 값을 변화시켜야함
price2 = df['가격'] < 800
```

```python
df.loc[price2, "가격"] = 0
```

```python
df[price2].loc[:,'가격']
```

- 지역에 강릉이라는 단어가 있는 컬럼 값을 바꾸기
